```python
win = []
for i in range(0,4010):
    if i%10 <5:
        win.append(final_df["radiant_win"][i])
    else:
        win.append(not final_df["radiant_win"][i])
```

```python
my_columns = ["localized_name","kills","assists","deaths","win"]
```

```python
clean_df = final_df.loc[:,my_columns]
```

```python
clean_df['kda']=(clean_df['kills']+clean_df['assists'])/(clean_df['deaths']+1)
```

```python
clean_df.to_csv("./cleaned_data.csv")
```

```python
heros_data_dict={}
for index in clean_df.index:
    record=clean_df.loc[index]
    hero=record['localized_name']
    kda=record['kda']
    win=record['win']
    #print(kda)
    if hero in heros_data_dict:
        heros_data_dict[hero]['count']+=1
        heros_data_dict[hero]['kda']+=kda
        heros_data_dict[hero]['win']+=1 if win else 0
    else:
        heros_data_dict[hero]={
            'count':1,
            'kda':kda,
            'win': 1 if win else 0
        }
heros_data_dict
        
    
```

```python
for k in heros_data_dict:
    heros_data_dict[k]['win_rate']=heros_data_dict[k]['win']/heros_data_dict[k]['count']
    heros_data_dict[k]['kda']=heros_data_dict[k]['kda']/heros_data_dict[k]['count']
```

```python
import re
heros_df=pd.read_csv('./heroes.csv')
heros_bp={}
count=0
for index in final_df.index:
    #print(index)
    record=final_df.loc[index]
    bps=record['picks_bans']
#     print(bps)
#     print('\n')
    p = re.compile(r'[{](.*?)[}]', re.S)
    parsed_bps=re.findall(p,bps)
    #print(parsed_bps)
    for bp in parsed_bps:
        parsed_bp=bp.strip().split(',')
#         print(parsed_bp)
        bp_dict={}
        for item in parsed_bp:
            #print(item)
            key,value=item.strip().split(':')
            #print(key)
            #print(value)
            bp_dict[key.strip()[1:-1]]=value.strip()
        #print(bp_dict)
        if bp_dict['hero_id'] in heros_bp:
            heros_bp[bp_dict['hero_id']]['ban']+=0 if bp_dict['is_pick']=='True' else 1
            heros_bp[bp_dict['hero_id']]['pick']+=1 if bp_dict['is_pick']=='True' else 0
        else:
            heros_bp[bp_dict['hero_id']]={
                'ban':0 if bp_dict['is_pick']=='True' else 1,
                'pick':1 if bp_dict['is_pick']=='True' else 0
            }
    count+=1
print(heros_bp)
print(count)
count/=10
# print(heros_df)
for i in heros_df.index:
    h=str(heros_df.loc[i]['id'])
    localized_name=str(heros_df.loc[i]['localized_name'])
    #print(h)
    #h=heros_df.loc[i]['id']
    if h in heros_bp:
        heros_bp[h]['ban']/=10
        heros_bp[h]['pick']/=10
        heros_bp[h]['other']=count-heros_bp[h]['ban']-heros_bp[h]['pick']
    else:
        heros_bp[h]={
            'ban':0,
            'pick':0,
            'other':count
        }
    heros_bp[h]['localized_name']=localized_name

heros_bp_with_name={}
for h in heros_bp:
    heros_bp_with_name[heros_bp[h]['localized_name']]=heros_bp[h]
print(heros_bp_with_name)
```

```python
#print(heros_data_df)
for h in heros_bp_with_name:
    if h in heros_data_dict:
        heros_data_dict[h]['pick']=heros_bp_with_name[h]['pick']
        heros_data_dict[h]['ban']=heros_bp_with_name[h]['ban']
        heros_data_dict[h]['other']=heros_bp_with_name[h]['other']
        heros_data_dict[h]['pick_rate']=heros_bp_with_name[h]['pick']/401
        heros_data_dict[h]['ban_rate']=heros_bp_with_name[h]['ban']/401
        heros_data_dict[h]['other_rate']=heros_bp_with_name[h]['other']/401
    else:
        heros_data_dict[h]={
            'count':0,
            'kda':'nan',
            'win':'nan',
            'win_rate':'nan',
            'pick':0,
            'ban':heros_bp_with_name[h]['ban'],
            'other':heros_bp_with_name[h]['other'],
            'pick_rate':0,
            'ban_rate':0,
            'other_rate':1.0
        }
print(heros_data_dict)   
    
    
```

```python
hero_dict={}
for idx in heros_df.index:
    hero_dict[heros_df.loc[idx]['localized_name']]={
        'attr':heros_df.loc[idx]['primary_attr'],
        'type':heros_df.loc[idx]['attack_type']
    }
hero_dict
```

```python
heros=[]
wins=[]
win_rates=[]
kdas=[]
counts=[]
picks=[]
bans=[]
others=[]
pick_rates=[]
ban_rates=[]
other_rates=[]
attrs=[]
types=[]
for h in heros_data_dict:
    heros.append(h)
    wins.append((heros_data_dict[h]['win']))
    win_rates.append(heros_data_dict[h]['win_rate'])
    kdas.append(heros_data_dict[h]['kda'])
    counts.append(int(heros_data_dict[h]['count']))
    picks.append(int(heros_data_dict[h]['pick']))
    bans.append(int(heros_data_dict[h]['ban']))
    others.append(int(heros_data_dict[h]['other']))
    pick_rates.append(heros_data_dict[h]['pick_rate'])
    ban_rates.append(heros_data_dict[h]['ban_rate'])
    other_rates.append(heros_data_dict[h]['other_rate'])
    attrs.append(hero_dict[h]['attr'])
    types.append(hero_dict[h]['type'])
heros_data_df=pd.DataFrame()
heros_data_df['hero']=heros
heros_data_df['win']=wins
heros_data_df['win_rate']=win_rates
heros_data_df['kda']=kdas
heros_data_df['count']=counts
heros_data_df['pick']=picks
heros_data_df['ban']=bans
heros_data_df['other']=others
heros_data_df['pick_rate']=pick_rates
heros_data_df['ban_rate']=ban_rates
heros_data_df['other_rate']=other_rates
heros_data_df['attr']=attrs
heros_data_df['type']=types
```
